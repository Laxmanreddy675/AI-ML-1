Question 1:
<br>Print ManOfMach count of each player in Hyderabad stadium

```python
pd.set_option('max_rows',10000)
```

```python
import numpy as np
import pandas as pd
data=pd.read_csv("cricket.csv",index_col=0)

from collections import Counter
NoOfManOfMatches=Counter(data['ManOfMach'])
count=pd.DataFrame(list(NoOfManOfMatches.items()),columns=["players","NoOfManOfMatches"])
count.sort_values("NoOfManOfMatches",ascending=False)
```

Question 2:<br>Plot a pie chart for the country_name column

```python
country=Counter(data['Country_Name'])
c=list(country.keys())
v=list(country.values())

import matplotlib.pyplot as plt
%matplotlib inline

plt.pie(v,labels=c,autopct='%0.2f%%',explode=[0.2,0,0],shadow=True,radius=2,startangle=45,colors=['blue','red','green'])
plt.show()
```
